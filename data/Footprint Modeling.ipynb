```python
from sklearn.model_selection import train_test_split
footprint = emissions['CarbonEmission']
```

```python
emission_t = emissions.drop('CarbonEmission', axis = 1)
```

```python
emission_t
```

```python
X_train, X_test, y_train, y_test = train_test_split(emission_t, footprint, test_size = 0.3)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import cross_val_score
```

```python
##base model
model = Lasso()
scores_r2 = cross_val_score(model, X_train, y_train,cv=10) # base on r2 from Lasso
scores_MAE = cross_val_score(model, X_train, y_train, scoring = 'neg_mean_absolute_error') # base on neg mean abs error

print('Training Set, Ave. R2 : {:.4f}'.format(np.mean(scores_r2)))
print('Training Set, Ave. MAE: {:.4f}'.format(np.mean(abs(scores_MAE)))) # 
```

```python
emissions['CarbonEmission'].mean()
```

```python
##finding best param with gridsearchCV
alpha_range = np.arange(0.01, 2, 0.01)
alpha_range
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param = {'alpha':alpha_range}
```

```python
model = Lasso(random_state = 0)

grid_search_r2 = GridSearchCV(model,param_grid=param,cv=5)
grid_search_r2.fit(X_train, y_train)

# score base on MAE
grid_search_MAE = GridSearchCV(model,param_grid=param,scoring ='neg_mean_absolute_error',cv=5)
grid_search_MAE.fit(X_train, y_train)



# results
print('Scoring R2')
print('Best R2 score   : ', grid_search_r2.best_score_)
print('Best parameters : ', grid_search_r2.best_params_)
print()


# results
print('Scoring MAE')
print('Best MAE score  : ', abs(grid_search_MAE.best_score_))
print('Best parameters : ', grid_search_MAE.best_params_)
```

```python
##Lasso Regression -> Lasso(alpha = 1.99), R^2 = 85.07%, avg margin of error = 269.96
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
reg = GradientBoostingRegressor(random_state = 0)

scores_r2 = cross_val_score(reg, X_train, y_train,cv=10) # base on r2 from gbr
scores_MAE = cross_val_score(reg, X_train, y_train, scoring = 'neg_mean_absolute_error') # base on neg mean abs error
```

```python
print('Training Set, Ave. R2 : {:.4f}'.format(np.mean(scores_r2)))
print('Training Set, Ave. MAE: {:.4f}'.format(np.mean(abs(scores_MAE)))) # 
```

```python
from sklearn.model_selection import RepeatedKFold
```

```python
model = GradientBoostingRegressor()
grid = dict()
grid['n_estimators'] = [10, 50, 100, 500]
grid['learning_rate'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
grid['subsample'] = [0.5, 0.7, 1.0]
grid['max_depth'] = [3, 7, 9]
# define the evaluation procedure
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# define the grid search procedure
grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv)
# execute the grid search
grid_result = grid_search.fit(X_train, y_train)
# summarize the best score and configuration
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_)) # s
```

```python
new_reg = GradientBoostingRegressor(learning_rate = 0.1, max_depth = 3, n_estimators = 500, subsample = 0.5)
```

```python
scores_r2 = cross_val_score(new_reg, X_train, y_train,cv=10) # base on r2 from gbr
scores_MAE = cross_val_score(new_reg, X_train, y_train, scoring = 'neg_mean_absolute_error') # base on neg mean abs error
```
